Вызовите документацию для класса RandomForestRegressor,
найдите информацию об атрибуте feature_importances_.


```python
?RandomForestRegressor
```

```python
#feature_importances_ : array of shape = [n_features] 
#    The feature importances (the higher, the more important the feature).
```

С помощью этого атрибута найдите сумму всех показателей важности,
установите, какие два признака показывают наибольшую важность.


```python
print(model.feature_importances_)
```

```python
model.feature_importances_.sum()
```

```python
#first
max_value_ind1=model.feature_importances_.argmax() 
print(max_value_ind1)
print("Значене:",model.feature_importances_[max_value_ind1])
```

```python
#second
max_value_ind2=0 
max_value=model.feature_importances_[max_value_ind2]
for i in range(model.n_features_):
    if max_value<model.feature_importances_[i] and i!=max_value_ind1:
        max_value=model.feature_importances_[i]       
        max_value_ind2=i 
print(max_value_ind2)
print("Значене:",max_value)
```

Разбейте X и y на тренировочный и тестовый наборы данных при помощи функции train_test_split, используя аргументы: test_size=0.3, random_state=100, stratify=y. У вас должны получиться объекты X_train, X_test, y_train и y_test.

```python
parameters = [{'n_estimators': [10, 15], 
'max_features': np.arange(3, 5), 
'max_depth': np.arange(4, 7)}]
 
```

Создайте модель GridSearchCV со следующими аргументами:
